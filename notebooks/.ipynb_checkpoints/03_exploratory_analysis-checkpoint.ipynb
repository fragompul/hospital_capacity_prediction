```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/hospital_spain_clean.csv", parse_dates=['date'])
df.tail()
```

```python
# Evolucin de hospitalizados y UCI
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='date', y='hosp_patients', label="Hospitalizados")
sns.lineplot(data=df, x='date', y='icu_patients', label="UCI")
plt.title("Evolucin de hospitalizados y UCI en Espaa")
plt.xlabel("Fecha")
plt.ylabel("Pacientes")
plt.legend()
plt.show()
```

```python
# Correlaciones
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```
